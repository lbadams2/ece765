# Computing Topological Ordering of a Directed Graph

This script illustrates how to obtain a topological ordering of a directed graph.

```python
import numpy as np
import networkx as nx
import matplotlib.pyplot as plt
import fncs as fncs
```

### Creating a Matrix and Checking its Initial Ordering

We load the data and use it for fitting the model.

```python
# Defining matrix
G = nx.DiGraph()
G.add_edges_from(
    [('A', 'B'), ('A', 'C'), ('D', 'B'), ('E', 'C'), ('E', 'F'),
     ('B', 'H'), ('B', 'G'), ('B', 'F'), ('C', 'G')])

# Plotting Graph
options = {
    'node_color': 'orange',
    'node_size': 400,
    'width': 2,
    'arrowstyle': '-|>',
    'arrowsize': 20,
}
nx.draw_networkx(G, arrows=True, **options)

# Checking Ordering
print('Topological Ordering is Satisfied: ',fncs.checkTopOrder(G.edges))
```

### Obtaining Topological Ordering

We apply our topological ordering next.

```python
Ord = fncs.compTopOrder(G.edges)
print('Ordering = ',Ord)
H = nx.relabel_nodes(G, Ord)

# Plotting Graph
nx.draw_networkx(H, arrows=True, **options)

# Checking Ordering
print('Topological Ordering is Satisfied: ',fncs.checkTopOrder(H.edges))
```
